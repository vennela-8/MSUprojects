```python
# proj1.py
#Canaeo Portage

# Constants
ROD_TO_METERS = 5.0292
METER_TO_FEET = 3.28084
MILE_IN_METERS = 1609.34
FURLONG_IN_RODS = 40
WALKING_SPEED_MPH = 3.1
MINUTES_PER_HOUR = 60

# Ask user for input
rods = float(input("Enter distance in rods: "))

# Calculations
meters = rods * ROD_TO_METERS
feet = meters * METER_TO_FEET
miles = meters / MILE_IN_METERS
furlongs = rods / FURLONG_IN_RODS
minutes = (miles / WALKING_SPEED_MPH) * MINUTES_PER_HOUR

# Output results
print("You entered:", rods, "rods")
print("In meters:", meters)
print("In feet:", feet)
print("In furlongs:", furlongs)
print("Minutes to walk that distance:",minutes)
```

```python
# proj2.py
#Richter Scale

# Function to calculate energy in joules and tons of TNT
def calculate_energy(richter):
    energy = 10 ** (1.5 * richter + 4.8)
    tnt = energy / (4.184 * 10 ** 9)
    return energy, tnt

# Predefined Richter values
print("Earthquake Energy Calculator\n")

print("Predefined Earthquake Magnitudes:")
richter_values = [1.0, 5.0, 9.1, 9.2, 9.5]

for r in richter_values:
    energy, tnt = calculate_energy(r)
    print("Richter:", r)
    print("Energy in joules:", energy)
    print("Equivalent in tons of TNT:", tnt)
    print()  # blank line for spacing

# User input
user_input = input("Enter a Richter scale value: ")
richter_value = float(user_input)

energy, tnt = calculate_energy(richter_value)

print("\nFor Richter value:", richter_value)
print("Energy in joules:", energy)
print("Equivalent in tons of TNT:", tnt)
```

```python
# proj3.py
#Windchill

# Function to calculate Wind Chill Temperature Index
def calculate_wind_chill(temp, wind_speed):
    wct_index = 35.74 + 0.6215 * temp - 35.75 * wind_speed*0.16 + 0.4275 * temp * wind_speed*0.16
    return wct_index

# Display heading
print("Wind Chill Temperature Index Calculator\n")

# Predefined values
predefined_values = [
    (10.0, 15.0),
    (0.0, 25.0),
    (-10.0, 35.0)
]

# Calculate and display for predefined values
for temp, wind in predefined_values:
    wct = calculate_wind_chill(temp, wind)
    print("Air Temp:", temp, "°F, Wind Speed:", wind, "MPH -> Wind Chill Index:", round(wct, 2))


# Get user input
user_temp = float(input("Enter air temperature in °F: "))
user_wind = float(input("Enter wind speed in MPH: "))

# Calculate user input result
user_wct = calculate_wind_chill(user_temp, user_wind)

# Display result
print("For Air Temp:", user_temp, "°F and Wind Speed:", user_wind, "MPH")
print("Wind Chill Temperature Index is:", round(user_wct, 2))
```

```python
#proj8.py
#Garden

import math

# Function to calculate the number of plants and cubic yards of soil
def garden_calculator():
    # Step 1: Get input values from the user
    side_length = float(input("Please enter the side length of your garden (in feet): "))
    plant_spacing = float(input("Please enter the distance between plants (in inches): "))
    flowerbed_depth = float(input("Please enter the depth for the flowerbeds (in feet): "))
    fill_depth = float(input("Please enter the depth for the fill areas (in feet): "))

    # Step 2: Convert plant spacing from inches to feet
    plant_spacing_feet = plant_spacing / 12  # convert inches to feet

    # Step 3: Calculate areas
    # Area of the entire garden (square)
    garden_area = side_length ** 2

    # Area of one triangular flowerbed (congruent isosceles triangles)
    triangle_area = 0.5 * (side_length / 2) * (side_length / 2)  # base * height / 2

    # Total area of the four triangular flowerbeds
    total_triangle_area = 4 * triangle_area

    # Area of the circular flowerbed (radius is half of side_length / 2)
    radius = side_length / 4
    circle_area = math.pi * (radius ** 2)

    # Total area of the flowerbeds (sum of triangles and circle)
    total_flowerbed_area = total_triangle_area + circle_area

    # Area of the fill areas (remaining area in the garden)
    fill_area = garden_area - total_flowerbed_area

    # Step 4: Calculate number of plants
    plant_area = plant_spacing_feet ** 2  # Area required per plant in square feet
    plants_per_triangle = triangle_area / plant_area
    plants_per_circle = circle_area / plant_area

    total_plants = (plants_per_triangle * 4) + plants_per_circle

    # Step 5: Calculate cubic yards of soil
    flowerbed_volume_cubic_feet = total_flowerbed_area * flowerbed_depth
    fill_area_volume_cubic_feet = fill_area * fill_depth

    total_flowerbed_soil_cubic_yards = flowerbed_volume_cubic_feet / 27  # Convert to cubic yards
    total_fill_soil_cubic_yards = fill_area_volume_cubic_feet / 27  # Convert to cubic yards

    # Step 6: Print the results
    print("\nSummary of your plant needs:")
    print(f"Total number of plants for the garden: {int(total_plants)}")
    print(f"Plants for each type of flowerbed (triangular): {int(plants_per_triangle)}")
    print(f"Plants for the circular flowerbed: {int(plants_per_circle)}")
    print(f"\nSummary of your soil needs:")
    print(f"Cubic yards of soil for flowerbeds: {total_flowerbed_soil_cubic_yards:.2f} cubic yards")
    print(f"Cubic yards of soil for fill areas: {total_fill_soil_cubic_yards:.2f} cubic yards")

# Run the program
garden_calculator()
```

```python
# proj10.py
# Number Puzzle: SplitNumber

def get_valid_number():
    while True:
        user_input = input("Enter a whole number: ")
        if user_input.isdigit():  # Check if input is a valid whole number
            print(f"Valid number entered: {user_input}")  # Debug print
            return user_input
        else:
            print("Invalid input. Please enter a valid whole number.")

def get_valid_piece_length(number):
    while True:
        user_input = input(f"Enter the number of digits in each piece (must be a divisor of {len(number)}): ")
        if user_input.isdigit():
            piece_length = int(user_input)
            if len(number) % piece_length == 0:
                print(f"Valid piece length entered: {piece_length}")  # Debug print
                return piece_length
            else:
                print(f"The number of digits in each piece must divide the length of the number ({len(number)}). Try again.")
        else:
            print("Invalid input. Please enter a valid number.")

def split_into_pieces(number, piece_length):
    pieces = [number[i:i + piece_length] for i in range(0, len(number), piece_length)]
    print(f"Pieces after split: {pieces}")  # Debug print
    return pieces

def check_increasing_order(pieces):
    # Convert the pieces to integers for comparison
    pieces_int = [int(piece) for piece in pieces]
    print(f"Converted pieces to integers: {pieces_int}")  # Debug print
    # Check if the list is in increasing order
    for i in range(1, len(pieces_int)):
        if pieces_int[i] <= pieces_int[i - 1]:
            return False
    return True

def main():
    # Step 1: Get the whole number input
    number = get_valid_number()

    # Step 2: Get the valid piece length
    piece_length = get_valid_piece_length(number)

    # Step 3: Split the number into pieces
    pieces = split_into_pieces(number, piece_length)

    # Step 4: Display the pieces
    print(f"The pieces are: {', '.join(pieces)}")

    # Step 5: Check if the pieces are in increasing order
    if len(pieces) == 1:
        print("There is only one piece, so it is considered in increasing order.")
    elif check_increasing_order(pieces):
        print("The pieces are in numerically increasing order.")
    else:
        print("The pieces are NOT in numerically increasing order.")

if __name__ == "__main__":  # Fixed the typo here
    main()
```

```python
# proj11.py
# CarRental

def get_classification_code():
    while True:
        # Ask the user for the classification code
        code = input("Enter the customer's classification code (B, D, W or Q to quit): ").upper()
        if code in ['B', 'D', 'W', 'Q']:  # Check if the input is valid
            return code
        else:
            print("Invalid classification code. Please enter 'B', 'D', 'W', or 'Q'.")

def get_number_input(prompt):
    while True:
        try:
            # Ask for an integer input and ensure it's a valid number
            return int(input(prompt))
        except ValueError:
            print("Invalid input! Please enter a valid number.")

def calculate_miles_driven(start, end):
    # Calculate the miles driven, accounting for rollover of the odometer
    if end >= start:
        return (end - start) / 10.0
    else:
        return ((1000000 - start) + end) / 10.0

def calculate_charge(code, days, miles):
    # Calculate the charge based on the classification code
    if code == 'B':  # Budget
        base_charge = 40 * days
        mileage_charge = 0.25 * miles
        total_charge = base_charge + mileage_charge
    elif code == 'D':  # Daily
        base_charge = 60 * days
        if miles / days > 100:  # Charge extra if average miles/day is above 100
            mileage_charge = 0.25 * (miles - 100 * days)
        else:
            mileage_charge = 0
        total_charge = base_charge + mileage_charge
    elif code == 'W':  # Weekly
        base_charge = 190 * ((days - 1) // 7 + 1)  # Full week charge
        if miles / 10 > 900:
            if miles / 10 > 1500:
                mileage_charge = 200 + 0.25 * (miles - 1500 * 10)
            else:
                mileage_charge = 100
        else:
            mileage_charge = 0
        total_charge = base_charge + mileage_charge
    else:
        total_charge = 0  # Default case (not needed since 'Q' will stop the loop)
    return total_charge

def main():
    while True:
        classification = get_classification_code()

        if classification == 'Q':  # Exit the program if 'Q' is entered
            print("Exiting the program.")
            break

        days = get_number_input("Enter the number of days the vehicle was rented: ")
        start_odometer = get_number_input("Enter the starting odometer reading: ")
        end_odometer = get_number_input("Enter the ending odometer reading: ")

        # Calculate miles driven
        miles = calculate_miles_driven(start_odometer, end_odometer)

        # Calculate total charge
        total_amount = calculate_charge(classification, days, miles)

        # Display the result
        print(f"\nRental Summary:")
        print(f"Classification code: {classification}")
        print(f"Days rented: {days}")
        print(f"Starting odometer reading: {start_odometer}")
        print(f"Ending odometer reading: {end_odometer}")
        print(f"Miles driven: {round(miles, 1)} miles")
        print(f"Total amount billed: ${total_amount:.2f}\n")

if __name__ == "__main__":  # Corrected the typo here
    main()
```

```python
#proj12.py
#Trig
#This program shows a menu with options to calculate:
# - Factorial of a number
# - Sine of an angle (in radians)
# - Cosine of an angle (in radians)

# Constant value of pi and epsilon for accuracy
PI = 3.141592653589793
EPSILON = 1.0e-8

# Show the menu options
def show_menu():
    print("\nMenu Options:")
    print("A. Display the value of the factorial of N.")
    print("B. Display the value of the sine of X.")
    print("C. Display the value of the cosine of X.")
    print("M. Display the menu of options.")
    print("Q. Exit from the program.")

# Calculate factorial of n (n!)
def factorial(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

# Reduce angle to range [0, 2π] to improve sine/cosine accuracy
def reduce_angle(x):
    return x % (2 * PI)

# Calculate sine using power series expansion
def calculate_sine(x):
    x = reduce_angle(x)
    term = x  # first term in the series
    total = 0.0
    k = 0

    while abs(term) >= EPSILON:
        total += term
        k += 1
        term = ((-1) ** k) * (x ** (2 * k + 1)) / factorial(2 * k + 1)

    return total

# Calculate cosine using power series expansion
def calculate_cosine(x):
    x = reduce_angle(x)
    term = 1.0  # first term in the series
    total = 0.0
    k = 0

    while abs(term) >= EPSILON:
        total += term
        k += 1
        term = ((-1) ** k) * (x ** (2 * k)) / factorial(2 * k)

    return total

# Main program loop
def main():
    show_menu()

    while True:
        user_choice = input("\nEnter your choice: ").strip().lower()

        if user_choice == 'a':
            n = int(input("Enter a positive integer N: "))
            if n < 0:
                print("Error: N must be a positive integer.")
            else:
                print(f"The factorial of {n} is {factorial(n)}")

        elif user_choice == 'b':
            x = float(input("Enter a real value X (in radians): "))
            print(f"The sine of {x} is {calculate_sine(x)}")

        elif user_choice == 'c':
            x = float(input("Enter a real value X (in radians): "))
            print(f"The cosine of {x} is {calculate_cosine(x)}")

        elif user_choice == 'm':
            show_menu()

        elif user_choice == 'q':
            print("Thank you! Exiting the program.")
            break

        else:
            print("Invalid option. Please choose again.")
            show_menu()

# Run the program
main()
```

```python
#proj13.py
#LatinSquares

def main():
    # Ask user for the order of the Latin Square
    order = int(input("Please enter the input of square: "))

    # Ask user for the top-left starting number
    start = int(input("Please Enter the top-left number (between 1 and " + str(order) + "): "))

    # Check if the top-left number is valid
    if start < 1 or start > order:
        print("Error: Top-left number must be between 1 and", order)
        return  # Exit the program early

    print("\nLatin Square of order", order, "starting with", start, ":\n")

    # Generate the Latin Square
    for row in range(order):
        for col in range(order):
            # Calculate the value using modular arithmetic
            value = ((start - 1 + row + col) % order) + 1
            print(value, end=" ")
        print()  # Move to the next row

# Run the main function
main()
```

```python
import re

def main():
    print("Welcome to the simple calculator!\n")

    while True:
        # Prompt user for an expression
        expression = input("Enter an expression (e.g., 4 + 5): ").strip()

        # Match expression using regex: number operator number
        match = re.fullmatch(r'(-?\d+(?:\.\d+)?)\s*([+\-*/])\s*(-?\d+(?:\.\d+)?)', expression)

        if not match:
            print("Error: Invalid input format. Use: number operator number (e.g., 4 + 5)")
            continue

        # Extract numbers and operator
        num1 = float(match.group(1))
        operator = match.group(2)
        num2 = float(match.group(3))

        # Perform calculation
        if operator == '+':
            result = num1 + num2
        elif operator == '-':
            result = num1 - num2
        elif operator == '*':
            result = num1 * num2
        elif operator == '/':
            if num2 == 0:
                print("Error: Division by zero is not allowed.")
                continue
            result = num1 / num2

        # Display result
        print("Result:", result)

        # Ask user if they want to continue
        again = input("Do you want to perform another calculation? (yes/no): ").strip().lower()
        if again not in ['yes', 'y']:
            print("Thank you for using the calculator. Goodbye!")
            break

# Run the program
main()
```

```python
#proj15.py
# Palindromic Integers

# Function to check if a number is a palindrome
def is_palindrome(n):
    return str(n) == str(n)[::-1]

# Function to reverse the digits of a number and return it as an integer
def reverse_number(n):
    return int(str(n)[::-1])

# Function to apply the 196-algorithm and determine if a number is a Lychrel number
def is_lychrel(n, max_iterations=60):
    for _ in range(max_iterations):
        n = n + reverse_number(n)
        if is_palindrome(n):
            return False  # Found a palindrome, not a Lychrel number
    return True  # Did not find a palindrome within max iterations, it's a Lychrel number

def main():
    # Prompt for the range of integers
    start = int(input("Enter the starting number of the range: "))
    end = int(input("Enter the ending number of the range: "))

    # Initialize counters
    natural_palindromes = 0
    non_lychrels = 0
    lychrels = 0

    # Iterate through the range and apply the 196-algorithm
    for num in range(start, end + 1):
        if is_palindrome(num):
            natural_palindromes += 1
        elif is_lychrel(num):
            lychrels += 1
            print(f"Lychrel number: {num}")
        else:
            non_lychrels += 1

    # Print the results
    print(f"\nSummary:")
    print(f"Natural Palindromes: {natural_palindromes}")
    print(f"Non-Lychrel Numbers: {non_lychrels}")
    print(f"Lychrel Numbers: {lychrels}")

# Correct entry point
if __name__ == "__main__":
    main()
```
